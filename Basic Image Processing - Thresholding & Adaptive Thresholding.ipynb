<center><h1>Introduction to Thresholding</h1></center>

```python
# import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read images in gray scale
img1 = cv2.imread('random1.jpg',0) # image size 320x390
img2 = cv2.imread('random2.jpg',0) # image size 1280x721
```

```python
# resize the image
img1 = cv2.resize(img1, (250, 250))
img2 = cv2.resize(img2, (700, 700))
```

## Thresholding 

### Binary Thresholding

```python
# here the threshold value is 100. So, any value below 100 will be made 0 and values above 100 will be made 255
ret, thresh1_img1 = cv2.threshold(img1, 100, 255, cv2.THRESH_BINARY)

# here the threshold value is 125. So, any value below 100 will be made 0 and values above 100 will be made 255
ret, thresh2_img1 = cv2.threshold(img1, 125, 255, cv2.THRESH_BINARY)

# here the threshold value is 170. So, any value below 100 will be made 0 and values above 100 will be made 255
ret, thresh3_img1 = cv2.threshold(img1, 170, 255, cv2.THRESH_BINARY)

# here the threshold value is 220. So, any value below 100 will be made 0 and values above 100 will be made 255
ret, thresh4_img1 = cv2.threshold(img1, 220, 255, cv2.THRESH_BINARY)
```

```python
# plot the images
titles = ['Original Gray Image','Threshold: 100', 'Threshold: 125', 'Threshold: 170', 'Threshold: 220']
images = [img1, thresh1_img1, thresh2_img1, thresh3_img1, thresh4_img1]

plt.figure(figsize=(10,6))
for i in range(5):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
```

### Inverted Binary Thresholding

```python
# here the threshold value is 100. So, any value below 100 will be made 255 and values above 100 will be made 0
ret, thresh1_img1 = cv2.threshold(img1, 100, 255, cv2.THRESH_BINARY_INV)

# here the threshold value is 125. So, any value below 100 will be made 255 and values above 100 will be made 0
ret, thresh2_img1 = cv2.threshold(img1, 125, 255, cv2.THRESH_BINARY_INV)

# here the threshold value is 170. So, any value below 100 will be made 255 and values above 100 will be made 0
ret, thresh3_img1 = cv2.threshold(img1, 170, 255, cv2.THRESH_BINARY_INV)

# here the threshold value is 220. So, any value below 100 will be made 255 and values above 100 will be made 0
ret, thresh4_img1 = cv2.threshold(img1, 220, 255, cv2.THRESH_BINARY_INV)
```

```python
# plot the images
titles = ['Original Gray Image','Threshold: 100', 'Threshold: 125', 'Threshold: 170', 'Threshold: 220']
images = [img1, thresh1_img1, thresh2_img1, thresh3_img1, thresh4_img1]

plt.figure(figsize=(10,6))
for i in range(5):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
```

### TRUNC Thresholding 

For a threshold value, any value below threshold will remain same and value above threshold will be made equal to the threshold value

```python
ret, thresh1_img1 = cv2.threshold(img1, 100, 255, cv2.THRESH_TRUNC)
ret, thresh2_img1 = cv2.threshold(img1, 125, 255, cv2.THRESH_TRUNC)
ret, thresh3_img1 = cv2.threshold(img1, 170, 255, cv2.THRESH_TRUNC)
ret, thresh4_img1 = cv2.threshold(img1, 220, 255, cv2.THRESH_TRUNC)
```

```python
# plot the images
titles = ['Original Gray Image','Threshold: 100', 'Threshold: 125', 'Threshold: 170', 'Threshold: 220']
images = [img1, thresh1_img1, thresh2_img1, thresh3_img1, thresh4_img1]

plt.figure(figsize=(10,6))
for i in range(5):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
```

## Adaptive Thresholding

Here, different threshold values are applied at different areas of image

### Mean  & Gaussian Thresholding 

```python
# global threshold value of 125 for entire image
ret, thresh1_img1 = cv2.threshold(img1, 125, 255, cv2.THRESH_BINARY)

# threshold values will change for every block
thresh2_img1 = cv2.adaptiveThreshold(img1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 2)
thresh3_img1 = cv2.adaptiveThreshold(img1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 9, 2)
```

```python
# plot the images
titles = ['Original Gray Image', 'Threshold: 125', 'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img1, thresh1_img1, thresh2_img1, thresh3_img1]

plt.figure(figsize=(10,6))
for i in range(4):
    plt.subplot(2, 2, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])
```

The <b>Gaussian Thresholding</b> looks more sharper than <b>Mean Thresholding</b>

```python
thresh1_img2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 9, 2)
thresh2_img2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 2)

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.imshow(img2, 'gray')
plt.title('Original Gray Image')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,2)
plt.imshow(thresh1_img2, 'gray')
plt.title('Adaptive Gaussian Image')
plt.xticks([])
plt.yticks([])

plt.subplot(2,2,3)
plt.imshow(thresh2_img2, 'gray')
plt.title('Adaptive Mean Image')
plt.xticks([])
plt.yticks([])
```
